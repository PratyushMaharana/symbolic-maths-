```python
import numpy as np
import scipy as sp
import sympy as smp
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

symbolic case ***first define all your symbol in sympy***

```python
x,a,b,c,=smp.symbols('x,a,b,c', real=True)
f= smp.exp(-a*smp.sin(x**2)) * smp.sin(b**x) * smp.log(c*smp.sin(x)**2/x)
```

***compute derivates using smp.diff(f,x) where f is a function you want to tale the derivative of and x is the variable you are taking the derivative with respect to***

```python
dfdx=smp.diff(f,x)
dfdx
```

 ***take the nth derivative d^n.f/dx^n by putting the optional argument at the end smp.diff(f,x,n)\***

```python
dx = smp.diff(f,x,4)
dx
```

```python
dx.subs([(x,4),(a,1),(b,2),(c,3)])     #subsituiting the values
```

```python
dx.subs([(x,4),(a,1),(b,2),(c,3)]).evalf()
```

```python
dx_f= smp.lambdify((x,a,b,c),dx)     #dx_f means is numerical number ,, that i give it a number  it return the number for me
```

```python
x=np.linspace(1,2,100)
```

```python
x=np.linspace(1,2,100)
y=dx_f(x , a=1,b=1,c=3)
#now plot
plt.plot(x,y, 'o--')
plt.ylabel('$d^4 f / dx^4$', fontsize=24)
plt.xlabel('$x$', fontsize=24)
```
